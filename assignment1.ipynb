##  Problem 4

```python
import time
```

```python
total_time = 0
LOOPS  = 1000000
THICKNESS = 0.00008
for i in range(LOOPS):
    start = time.time()
    folded_thickness = THICKNESS*(2**43)
    elapsed_time = time.time() - start
    total_time += elapsed_time
print("Time using exponentiation arithmetic operators: {}[ns]".format((total_time/LOOPS)*1000000000))
```

```python
total_time = 0
LOOPS  = 1000000
for i in range(LOOPS):
    start = time.time()
    folded_thickness = 0.00008
    for i in range(43):
        folded_thickness = folded_thickness*2
    elapsed_time = time.time() - start
    total_time += elapsed_time
print("Time using for statement: {}[ns]".format((total_time/LOOPS)*1000000000))
```

Therefore, using exponentiation arithmetic operators is better
